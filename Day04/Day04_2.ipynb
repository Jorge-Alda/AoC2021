```python
with open('input') as f:
    data = f.read()
```

```python
inputlines = data.split('\n')
```

```python
numbers = inputlines[0].split(',')
```

```python
class BingoCard():

    def __init__(self, lines):
        self.dim = len(lines)
        self.rows = [0,]*self.dim
        self.cols = [0,]*self.dim
        self.numbers = {}
        self.won = False
        for i, l in enumerate(lines):
            for j in range(self.dim):
                n = l[3*j:3*j+2]
                if n[0] == ' ':
                    n = n[1]
                self.numbers.update({n:[i, j, False]})

    def play_number(self, n):
        if n in self.numbers.keys():
            self.rows[self.numbers[n][0]] += 1
            self.cols[self.numbers[n][1]] += 1
            self.numbers[n][2] = True

    def check(self):
        for i in self.cols:
            if i == self.dim:
                self.won = True
                return True
        for i in self.rows:
            if i == self.dim:
                self.won = True
                return True
        return False

    def sum_unmarked(self):
        s = 0
        for n, v in self.numbers.items():
            if not v[2]:
                s += int(n)
        return s
```

```python
NCards = len(inputlines)//6-1
Cards = []
for i in range(NCards):
    Cards.append(BingoCard(inputlines[i*6+2:i*6+7]))
```

```python
for n in numbers:
    for C in Cards:
        if not C.won:
            C.play_number(n)
            if C.check():
                score = int(n) * C.sum_unmarked()
                print(score)
```
